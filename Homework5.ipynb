##Step 2:
Create an empty list named request
  * Use function called build_request(), to ask buyer for item input
  * Use while True loop and break statement to allow for multiple requests until users enters "submit"
  * If an invalid item is requested print "Item not in catalog" and ask again
  * Append items to request list

```python
def build_request(catalog):
  request=[]
  #empty list that will take items from user input if it is in the catalog dictionary
  while True:
    item=input("Enter an item to request (type 'submit' to finish): ").strip().lower()
    if item=="submit":
      break
    if item not in catalog:
      print("Item not in catalog")
      #if requested items are not in catalog "Item not in catalog" will print
    else:
      request.append(item)
      print(f"Added {item} to purchase request")
  return request
#at the end it will return the items requested
```

#Step 3
Calculate the subtotal. Define a function calculate_subtotal(), that:
* Loops through request list.
* Retrieves each item's unit price from the catalog.
* Calculate the subtotal and print it.

```python
def calculate_subtotal(request,catalog): #function will loop through request list and catalog
  subtotal=0.0
  print(f"Purchase Request Summary:")
  for item in request:
    price=catalog[item] #based on the items requested, this will capture the corresponding price from the catalog dictionary
    print(f"  {item}: ${price:.2f}") #prints the item and its price
    subtotal+=price #adds the price of each item to the subtotal
  print("")
  print(f"Subtotal: ${subtotal:.2f}") #prints the subtotal off all the items requested
  return subtotal
```
